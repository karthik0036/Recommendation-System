```python
features = ['keywords','cast','genres','director']
```

```python
##Step 3: Create a column in DF which combines all selected features
for feature in features:
    imdb[feature] = imdb[feature].fillna('')
    
def combine_features(row):
    try:
        return row['keywords'] +" "+row['cast']+" "+row["genres"]+" "+row["director"]
    except:
        print ("Error:", row)

imdb["combined_features"] = imdb.apply(combine_features,axis=1)

imdb["combined_features"].head()
```

```python
##Step 4: Create count matrix from this new combined column
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer()
count_matrix=cv.fit_transform(imdb["combined_features"])
print(count_matrix.toarray())
```

```python
##Step 5: Compute the Cosine Similarity based on the count_matrix

from sklearn.metrics.pairwise import cosine_similarity
cos_sim= cosine_similarity(count_matrix)
#print(similarity_score)
movie_user_likes = "The Avengers"
```

```python
## Step 6: Get index of this movie from its title

def get_title_from_index(index):
    return imdb[imdb.index == index]["title"].values[0]

def get_index_from_title(title):
    return imdb[imdb.title == title]["index"].values[0]


movie_index = get_index_from_title(movie_user_likes)
similar_movies =  list(enumerate(cos_sim[movie_index]))

## Step 7: Get a list of similar movies in descending order of similarity score
sorted_similar_movies = sorted(similar_movies,key=lambda x:x[1],reverse=True)
```

```python
## Step 8: Print titles of first 50 movies
i=0
print("Top  similar movies like "+movie_user_likes+" are:\n")
for element in sorted_similar_movies:
    
    print(get_title_from_index(element[0]))
    i=i+1
    if i>50:
        break
```

```python
sort_by_average_vote = sorted(sorted_similar_movies,key=lambda x:imdb["vote_average"][x[0]],reverse=True)
print(sort_by_average_vote)
```
